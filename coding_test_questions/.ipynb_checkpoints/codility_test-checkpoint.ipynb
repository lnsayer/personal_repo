```python
def solution(N):
    if N % 2 == 1:
        # If N is odd, we can use one letter repeated N times
        return 'a' * N
    else:
        # If N is even, we can use one letter N-1 times and another letter once
        return 'a' * (N - 1) + 'b'
```

```python
def solution3(X, Y):
    # Initialize the area under the curve
    auroc = 0.0
    
    # Loop through the points and calculate the area using the trapezoidal rule
    for i in range(1, len(X)):
        # Calculate the width of the trapezoid
        width = X[i] - X[i-1]
        # Calculate the average height of the trapezoid
        height = (Y[i] + Y[i-1]) / 2.0
        # Add the area of the trapezoid to the total area
        auroc += width * height
    
    return auroc
```
